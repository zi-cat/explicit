# notebook outline


In this notebook, I determine the normality of the score distribution of the stories. Following this, I perform statistical testing to determine whether or not there is a difference between male and female-authored stories' scores.


    valence is the positive--negative or pleasure--displeasure dimension;
    arousal is the excited--calm or active--passive dimension; and
    dominance is the powerful--weak or 'have full control'--'have no control' dimension.


# determining if data are normal

# freq plots

```python
import pandas as pd
import numpy as np
```

```python
master_df = pd.read_csv("master_df.csv")
```

```python
def reject_outliers(data, m = 2.):
    d = np.abs(data - np.median(data))
    mdev = np.median(d)
    s = d/mdev if mdev else 0.
    return data[s<m]
```

```python
from scipy.stats import norm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

datos = master_df[master_df.sex=='F']['norm_emo_score'].dropna()
datos = reject_outliers(datos)

# best fit of data
(mu, sigma) = norm.fit(datos)

# the histogram of the data
n, bins, patches = plt.hist(datos, 60, normed=1, facecolor='green', alpha=0.75)

# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=2)

#plot
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ emo(F):}\ \mu=%.3f,\ \sigma=%.3f$' %(mu, sigma))
plt.grid(True)

plt.show()
```
